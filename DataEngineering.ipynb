# DATA ENGINEERING

WHAT IS DATA SCIENCE ?
* Data Science is all about using data to make business decisions.One of the
ways we can make business decisions is by using a technique like Machine
Learning.So Machine Learning is a technique and Data science is the idea of 
using a data and converting it into something useful for a product or a 
business.

WHAT IS DATA ANALYSIS?
* Data Analysis is a subset of data science that allows us to analyze the data 
that we have

WHAT IS A DATA ENGINEER ?
* Data Engineer is a person who take data from all of the data points incoming.
Lets say for a company,the data coming from users,security cameras,from their 
website,etc.A data engineer takes all of this information,produces it and 
maintains it in certain types of databases or certain types of computers so the
data can be accessed for the business purpose.

# FOUR DIFFERENT TYPES OF DATA

Structured Data
* This is data that is organized well for us to understand.Often stored usually
in a table or a matrix or in pandas DataFrame.They usually come from 
relational Databases like MySQL,PostGres,etc

Semi-Structured Data
* They are still structured data but often occurs in something like an Excel 
CSV or json or XML.

UnStructured Data
* They are not usually in simple formats that we can manipulate and analyze
easily.Usaually like PDFs.

Binary Data
* These are things such as audio files,video files,images that occurs in binary
and computer can easily understand them but it is hard for us to manipulate
and categorize them.

DATA MINING
* Data Mining means preprocessing and extracting some knowledge from the data.

Big Data
* Its the data which is so big that we can't have on one computer.Its so big 
that we have it running on cloud computing or multiple computers such as AWS
Azure and Google cloud because they have lots of storage to store that data
